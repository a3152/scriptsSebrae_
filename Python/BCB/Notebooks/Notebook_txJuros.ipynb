```python
from bcb import TaxaJuros
```

```python
em = TaxaJuros()
em.describe()
```

```python
em.describe("TaxasJurosMensalPorMes")
em.describe("ParametrosConsulta")
em.describe("TaxasJurosDiariaPorInicioPeriodo")
em.describe("ConsultaUnificada")
em.describe("ConsultaDatas")
```

```python
import pandas as pd

ep = em.get_endpoint('TaxasJurosMensalPorMes').query().collect()
ep.groupby("Modalidade").agg({"TaxaJurosAoMes": "median"})
```

```python
ep.groupby("InstituicaoFinanceira").agg({"TaxaJurosAoMes": "median"})
```

```python
ep2 = em.get_endpoint('ParametrosConsulta').query().collect()
ep2.groupby("segmento").agg({"codigoModalidade": "count"})
```

```python
ep3 = em.get_endpoint('ConsultaUnificada').query().collect()
ep3.groupby("Modalidade").agg({"TaxaJurosAoMes": "median"})
```
